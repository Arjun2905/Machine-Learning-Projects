```python
updated_data = movies_df.drop(['title_x','poster_path','wiki_link','title_y','runtime','story','tagline','release_date'],axis=1)
```

```python
updated_data.wins_nominations.fillna(value=0,inplace=True)
updated_data.actors.fillna('',inplace=True)
updated_data.head()
```

```python
updated_data.info()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=2,max_features=None,strip_accents='unicode',analyzer='word',token_pattern=r'\w{1,}',stop_words='english',ngram_range=(1,3))
```

```python
tfv_matrix = tfv.fit_transform(updated_data['summary'])
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

sig = sigmoid_kernel(tfv_matrix,tfv_matrix)
```

```python
sig[0]
```

```python
indexes = pd.Series(updated_data.index,index=updated_data['original_title']).drop_duplicates()
```

```python
indexes
```

```python
def giv_rec(title,sig=sig):
  idx = indexes[title]
  sig_score = list(enumerate(sig[idx]))
  sig_score = sorted(sig_score, key=lambda x: x[1], reverse=True)
  sig_score = sig_score[1:11]
  movie_idx = [i[0] for i in sig_score]

  return updated_data['original_title'].iloc[movie_idx]
```
